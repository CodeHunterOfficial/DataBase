```python
import sqlite3

# Создание соединения с базой данных
conn = sqlite3.connect('university.db')
c = conn.cursor()

# Создание таблицы "Студенты"
c.execute("""CREATE TABLE Students (
             id INTEGER PRIMARY KEY,
             name TEXT,
             age INTEGER,
             email TEXT,
             course INTEGER
           )""")

# Создание таблицы "Преподаватели"
c.execute("""CREATE TABLE Professors (
             id INTEGER PRIMARY KEY,
             name TEXT,
             department TEXT,
             subject TEXT
           )""")

# Создание таблицы "Курсы"
c.execute("""CREATE TABLE Courses (
             id INTEGER PRIMARY KEY,
             name TEXT,
             credits INTEGER,
             professor_id INTEGER,
             FOREIGN KEY (professor_id) REFERENCES Professors(id)
           )""")

# Создание таблицы "Оценки"
c.execute("""CREATE TABLE Grades (
             id INTEGER PRIMARY KEY,
             student_id INTEGER,
             course_id INTEGER,
             grade REAL,
             FOREIGN KEY (student_id) REFERENCES Students(id),
             FOREIGN KEY (course_id) REFERENCES Courses(id)
           )""")

# Создание таблицы "Кампусы"
c.execute("""CREATE TABLE Campuses (
             id INTEGER PRIMARY KEY,
             name TEXT,
             location TEXT
           )""")

# Создание таблицы "Общежития"
c.execute("""CREATE TABLE Dormitories (
             id INTEGER PRIMARY KEY,
             name TEXT,
             capacity INTEGER,
             campus_id INTEGER,
             FOREIGN KEY (campus_id) REFERENCES Campuses(id)
           )""")

# Создание таблицы "Зачисления"
c.execute("""CREATE TABLE Enrollments (
             id INTEGER PRIMARY KEY,
             student_id INTEGER,
             course_id INTEGER,
             semester TEXT,
             FOREIGN KEY (student_id) REFERENCES Students(id),
             FOREIGN KEY (course_id) REFERENCES Courses(id)
           )""")

# Сохранение изменений и закрытие соединения
conn.commit()
conn.close()
```

В этом коде мы создали следующие таблицы:

* Students: Содержит информацию о студентах.
* Professors: Содержит информацию о преподавателях.
* Courses: Содержит информацию о курсах, включая связь с преподавателями.
* Grades: Содержит информацию об оценках студентов, включая связь со студентами и курсами.
* Campuses: Содержит информацию о кампусах.
